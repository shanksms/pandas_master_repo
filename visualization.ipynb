# General matplotlib tips

**importing matplotlib**

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

**Setting style**

```python
plt.style.use('classic')
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```

# Simple line plots

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

**For all matplotlib plots , we start by creating a figure and an axes.**  


```python
fig = plt.figure()
ax = plt.axes()
```

**In matplotlib, the figure can be thought of as a single container that contains all the objects representing axes, graphics, text and labels. The axes is what we see above: a bounding box with tix and labels, which will eventually conatain the plot elements that make up our visualization.**

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
```

**Alternatively we can use plt interface and let the figure and axes be created for us in the background.**

```python
plt.plot(x, np.sin(x))
```

**if we want to create a single figure with multiple lines, we can simply call the plot function multiple times**

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```
