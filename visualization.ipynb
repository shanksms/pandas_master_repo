# General matplotlib tips

**importing matplotlib**

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

**Setting style**

```python
plt.style.use('classic')
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
```
